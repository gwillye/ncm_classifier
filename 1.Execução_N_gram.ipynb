# Análise Descritiva e NLP - Capítulo 44 (Versão Integrada)

Este notebook contém a análise descritiva e a implementação de técnicas de NLP (N-grams) aplicadas diretamente ao dataset de itens de nota fiscal do Capítulo 44.

## 1. Configuração Inicial e Carregamento de Dados

```python
import pandas as pd
import re
import os
from sklearn.feature_extraction.text import TfidfVectorizer

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

FILE_PATH = 'cap_44.csv'

try:
    df_pandas = pd.read_csv(FILE_PATH, sep='\t', decimal='.')
    df_pandas['prod_xprod_norm'] = df_pandas['prod_xprod'].astype(str).str.upper()
    print(f"Total de linhas carregadas: {len(df_pandas)}")
except Exception as e:
    print(f"Erro ao carregar dados: {e}")
```

## 2. Implementação de N-grams (Dataset Real)

Nesta seção, aplicamos N-grams de palavras e caracteres diretamente sobre a coluna `prod_xprod_norm` do dataset principal.

### 2.1 Word N-grams (Bigrams e Trigrams)
Captura sequências de 2 e 3 palavras para entender o contexto (ex: 'JANELA MADEIRA').

```python
# Configurando o vetorizador para Word N-grams (2 a 3 palavras)
# Limitamos a 1000 features para visualização rápida, mas pode ser aumentado
vectorizer_word = TfidfVectorizer(ngram_range=(2, 3), max_features=1000)
tfidf_word = vectorizer_word.fit_transform(df_pandas['prod_xprod_norm'])

print("Exemplos de Word N-grams (Bigrams/Trigrams) encontrados no dataset:")
print(vectorizer_word.get_feature_names_out()[:20])
```

### 2.2 Character N-grams (3-grams a 5-grams)
Captura a estrutura interna das palavras, sendo ideal para lidar com erros de digitação e abreviações técnicas.

```python
# Configurando o vetorizador para Character N-grams (3 a 5 caracteres)
# O parâmetro 'char_wb' cria n-grams apenas dentro dos limites das palavras
vectorizer_char = TfidfVectorizer(analyzer='char_wb', ngram_range=(3, 5), max_features=1000)
tfidf_char = vectorizer_char.fit_transform(df_pandas['prod_xprod_norm'])

```

```python
print("Exemplos de Character N-grams encontrados no dataset:")
print(vectorizer_char.get_feature_names_out()[:1000])
```

### 2.3 Análise Filtrada: N-grams para o Token 'JANELA'
Filtramos o dataset para analisar especificamente os padrões que acompanham a palavra 'JANELA'.

```python
token_filtro = 'JANELA'
df_filtrado = df_pandas[df_pandas['prod_xprod_norm'].str.contains(token_filtro, na=False)]

if not df_filtrado.empty:
    # Word N-grams para o filtro
    vec_word_filt = TfidfVectorizer(ngram_range=(2, 3), max_features=500)
    tfidf_word_filt = vec_word_filt.fit_transform(df_filtrado['prod_xprod_norm'])
    
    # Character N-grams para o filtro
    vec_char_filt = TfidfVectorizer(analyzer='char_wb', ngram_range=(3, 5), max_features=500)
    tfidf_char_filt = vec_char_filt.fit_transform(df_filtrado['prod_xprod_norm'])
    
    print(f"Análise para o token: {token_filtro}")
    print(f"Total de itens encontrados: {len(df_filtrado)}")
    print("\nTop Word N-grams no contexto de JANELA:")
    print(vec_word_filt.get_feature_names_out()[:50])
else:
    print(f"Nenhum item encontrado com o token '{token_filtro}'.")
```
